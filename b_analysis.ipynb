```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('breast_cancer_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# remove 'id' because just a number and remove 'Unnamed' because only has Null values
df.drop(['id', 'Unnamed: 32'], inplace=True, axis=1)
```

```python
df.head()
```

```python
cols = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 
        'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 
        'concave points_mean', 'symmetry_mean',]

real_df = df[cols].copy()
real_df.head()
```

```python
sns.countplot(x='diagnosis', data=real_df, palette='Set1')
```

```python
sns.pairplot(real_df, hue='diagnosis', palette='Set1')
```

```python
# radius, perimeter and area are pretty much similar to each other
# remove area and perimeter for easier calculations?
real_df.drop(['perimeter_mean', 'area_mean'], inplace=True, axis=1)
real_df.head()
```

### Predictions

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X = real_df.drop('diagnosis', axis=1)
y = real_df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

#### Logistic Regression

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

#### Random Forest

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
print(confusion_matrix(y_test,rfc_pred))
print(classification_report(y_test,rfc_pred))
```
